```
# importing libraries to be used
import pandas as pd
```

```
# loading the datasets for MTN

# loading the first dataset cells_geo which is a csv file
cells_df = pd.read_csv('cells_geo.csv', index_col=0, sep=';')
cells_df
```

```
# loading the excel files dataset for Day 1

df1 = pd.read_csv('Telcom_dataset.csv')
df1
```

```
# loading the excel file dataset for Day 2

df2 = pd.read_csv('Telcom_dataset2.csv')
df2
```

```
# loading the excel file dataset for Day 3

df3 = pd.read_csv('Telcom_dataset3.csv')
df3
```

```
# merging dataframes
DF_telcom = pd.concat([df1,df2,df3], axis=0)
DF_telcom
```

```
# removing the missing values from my dataset
# here I decided to drop three columns that I did not need
DF_telcom.drop(['CELL_ON_SITE','COUNTRY_A','COUNTRY_B'], axis = 1, inplace=True)
DF_telcom
```

```
# I decided to merge the data frame DF_telcom with the cells_geo dataset

Telcom_new = pd.merge(left=cells_df,right=DF_telcom, left_on='SITE_CODE', right_on='SITE_ID')
Telcom_new
```

```
# checking for missing values 
Telcom_new.isnull()
```

```
# finding the top 10 most used city for 3 days

Telcom_new.sort_values(by='DATE_TIME',ascending=0).head(10)
```

```
# group values by cities 
#
#city_group
Telcom_new.sort_values(by='VALUE',ascending=0).head(10)
```

```
Telcom_new['VALUE'].groupby(Telcom_new['VILLES']).mean().head(10)
```

```
# regions from where the most used cities come from
Telcom_new['VILLES'].groupby(Telcom_new['REGION']).count()
```

```
Telcom_new.groupby(['REGION', 'VILLES']).size().head(50)
```

```
#Telcom_new['VILLES'].groupby(['REGION']).value_counts()
GROUP = Telcom_new.groupby('REGION')['VILLES'].value_counts().head(59)  
GROUP                          
```
